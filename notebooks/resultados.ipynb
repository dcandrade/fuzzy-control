```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('altura_sem_vazao.txt').set_index('tempo')
df.columns = ['sem vazão']
plot = df.plot();
plot.set_xlabel('Tempo (s)')
plot.set_ylabel('Altura (m)')
```

```python
df_vazao = pd.read_csv('altura_com_vazao.txt').set_index('tempo')
df_vazao.columns = ['com vazão']
plot = df_vazao.plot();
plot.set_xlabel('Tempo (s)')
plot.set_ylabel('Altura (m)')
```

```python
plot = pd.concat([df, df_vazao], axis=1).plot()
plot.set_xlabel('Tempo (s)')
plot.set_ylabel('Altura (m)')
```

### resultados

```python
t1 = pd.read_csv('var_1.txt').set_index('tempo')
t1.columns = ['Temperatura', 'Altura', 'Vazão','Vazão Ideal', 'Temperatura Ideal']
t1.head()
```

```python
t1[['Vazão', 'Vazão Ideal']].plot()
plt.xlabel('Tempo (s)')
plt.ylabel('Vazão (m^3/s)')
plt.title('Evolução da Vazão na Variação #1')
plt.savefig('variacao-parametros/vazao1.eps', format='eps')
```

```python
plot = t1[['Temperatura', 'Temperatura Ideal']].plot()
plt.xlabel('Tempo (s)')
plt.ylabel('Temperatura (ºCelsius)')
plt.title('Evolução da Temperatura na Variação #1')
plt.savefig('variacao-parametros/temperatura1.eps', format='eps')
```

### variação dos conjuntos fuzzy para o caso 2

```python
tf2 = pd.read_csv('var_2_set.txt').set_index('tempo')
tf2.columns = ['Temperatura (Nova Fuzzificação)', 'Altura', 'Vazão (Nova Fuzzificação)','Vazão Ideal', 'Temperatura Ideal']
tf2.head()
```

```python
tf2['Temperatura (Fuzzificação Original)'] = t2['Temperatura']
tf2['Vazão (Fuzzificação Original)'] = t2['Vazão']

tf2.head()
```

```python
tf2[['Vazão (Nova Fuzzificação)', 'Vazão Ideal', 'Vazão (Fuzzificação Original)']].plot()
plt.xlabel('Tempo (s)')
plt.ylabel('Vazão (m^3/s)')
plt.title('Evolução da Vazão na Variação #2 Alterando os Conjuntos Fuzzy')
plt.savefig('variacao-conjuntos/vazao.eps', format='eps')
```

```python
plot = tf2[['Temperatura (Nova Fuzzificação)', 'Temperatura Ideal', 'Temperatura (Fuzzificação Original)']].plot()
plt.xlabel('Tempo (s)')
plt.ylabel('Temperatura (ºCelsius)')
plt.title('Evolução da Temperatura na Variação #2 Alterando os Conjuntos Fuzzy')
plt.savefig('variacao-conjuntos/temperatura.eps', format='eps')
```
